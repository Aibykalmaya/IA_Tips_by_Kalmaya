### Guide Professionnel : Créer un Notebook et Sélectionner un Environnement Virtuel comme Noyau dans VS Code

Suivez ces étapes pour créer un environnement virtuel, l'activer et l'utiliser comme noyau Jupyter dans un notebook sous VS Code.

---

#### 1. Créer un Environnement Virtuel
Dans un terminal, naviguez vers le répertoire où vous souhaitez créer l'environnement virtuel, puis exécutez :

```bash
PS C:\Users\lenovo> cd Desktop\Pojets_pro2025\IA_tips
```

```bash
PS C:\Users\lenovo\Desktop\Pojets_pro2025\IA_tips> python -m venv MyEnv
```

> **Note** : Remplacez `MyEnv` par le nom que vous souhaitez pour votre environnement virtuel.

---

#### 2. Accéder à l'Environnement Virtuel
Naviguez dans le dossier de l'environnement virtuel nouvellement créé avec :

```bash
PS C:\Users\lenovo\Desktop\Pojets_pro2025\IA_tips> cd MyEnv
```

---

#### 3. Activer l'Environnement Virtuel
Activez l'environnement virtuel avec la commande suivante :

```bash
PS C:\Users\lenovo\Desktop\Pojets_pro2025\IA_tips\MyEnv> .\Scripts\activate
```

> **Important** : Assurez-vous d'être dans le dossier de l'environnement (`MyEnv`) pour que cette commande fonctionne correctement. comme ici dans mon cas:
```bash
PS C:\Users\lenovo\Desktop\Pojets_pro2025\IA_tips\MyEnv>
```
 Une fois activé, vous verrez le nom de l'environnement, comme `(MyEnv)`, en début de ligne dans votre terminal:
 ```bash
(MyEnv) PS C:\Users\lenovo\Desktop\Pojets_pro2025\IA_tips\MyEnv> 
```


---

#### 4. Installer `ipykernel` dans l'Environnement Virtuel
Pour permettre à Jupyter de reconnaître cet environnement, installez `ipykernel` :

```bash
(MyEnv) PS C:\Users\lenovo\Desktop\Pojets_pro2025\IA_tips\MyEnv> pip install ipykernel
```

> **Remarque** : Assurez-vous que votre environnement virtuel est activé avant d'exécuter cette commande, sinon `ipykernel` sera installé globalement.

---

#### 5. Ajouter l'Environnement Virtuel aux Noyaux Jupyter
Enregistrez votre environnement virtuel comme noyau Jupyter en exécutant la commande suivante :

```bash
(MyEnv) PS C:\Users\lenovo\Desktop\Pojets_pro2025\IA_tips\MyEnv> python -m ipykernel install --user --name=MyEnv --display-name "Python (MyEnv)"
```

> **Explications** :
> - `--name=MyEnv` : Définit le nom interne de l'environnement (vous pouvez le personnaliser).
> - `--display-name "Python (MyEnv)"` : Nom qui apparaîtra dans la liste des noyaux de Jupyter.

---

#### 6. Sélectionner l'Environnement Virtuel comme Noyau dans VS Code
- Si vous avez fait tout ce qui précede sur VS Code, fermer le et réouvrir VS Code à nouveau
- deplacer vous à nouveau jusqu'au niveau de votre environement virtuelle et activer le (voir ce qui précède), vous aurez un truc comme ça:
```bash
(MyEnv) PS C:\Users\lenovo\Desktop\Pojets_pro2025\IA_tips\MyEnv>
```
- Ouvrez votre notebook dans VS Code ou creer un nouveau notebook (ctrl + shift + p) selecter "creat a new jupyter notebook".
- Cliquez sur **"Select Kernel"** en haut à droite de l'interface du notebook.
- Choisissez **Python (MyEnv)** dans la liste des noyaux pour utiliser cet environnement.

![voir l'image]("select_kernel.png")

---

> **Vérification** : Pour confirmer que vous utilisez le bon environnement, exécutez cette commande dans une cellule de votre notebook :

```python
import sys
print(sys.executable)
```

Cela affichera le chemin vers l'interpréteur Python de votre environnement virtuel, confirmant qu'il est bien activé. Vous êtes maintenant prêt à utiliser VS Code avec votre environnement virtuel comme noyau pour les notebooks Jupyter.

````python
import sys
print(sys.executable)
````
